1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# Define the feature transformation function
def get_education_type(education):
    education = education.split()
    output = ''
    for elem in education:
        if elem != 'образование':
            output += ' ' + elem
        else:
            return output

        
# Make new feature        
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(get_education_type)

# check the result
display(hh_df['Образование'].unique()) 

hh_df['Образование'].value_counts()
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Define get the work experience feature function
def get_experience(input):
    if input in ['', ' ', 'Не указано'] or input is np.nan:  # Return NaN if haven't value
        return np.nan
    
    input = input.split()[2::]
    output = 0
    lst_years = ['лет', 'год', 'года']  # Rule for years
    lst_months = ['месяцев', 'месяца', 'месяц']  # Rule for months
    
    if input[1] in lst_months:
        output += int(input[0])
    elif input[1] in lst_years:
        output += int(input[0])*12
        if input[3] in lst_months:
            output += int(input[2])
            
    return output


# Make new feature       
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)

# Median work experience
display(hh_df['Опыт работы (месяц)'].median()) 

hh_df.drop(['Опыт работы'], axis= 1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', \
                  'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]


# Define get the city feature function
def get_city(input):
    city = input.split()[0]
   
    if city in million_cities:
        return 'город-миллионник'
    elif city in ['Москва', 'Санкт-Петербург']:
        return city
    else:
        return 'другие'


# Define get the relocate feature function
def get_relocate(input):
    if 'не готов к переезду' in input or 'не готова к переезду' in input:
        return False
    elif 'хочу переехать' in input:
        return True
    else:
        return True

    
# Define get trip ready feature function
def get_trip_ready(input):
    start = input.find('переезду')      # Find the beginning of useful 
    input = input[start + 10::]         # data block

    if 'не готов'in input or 'не готова' in input:
        return False
    elif 'готов'in input or 'готова' in input:
        return True
    else:
        return True
    
        
# Make new features        
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(get_relocate)
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(get_trip_ready)

# Cities share
display(hh_df['Город'].value_counts(normalize=True)*100) 

mask = (hh_df['Готовность к переезду'] & hh_df['Готовность к командировкам'])

display((hh_df[mask].shape[0]/hh_df.shape[0])*100)

hh_df.drop(['Город, переезд, командировки'], axis= 1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Define get occupation feature function
def get_occupation_full(input):
    return True if 'полная занятость' in input else False


# Define get occupation feature function
def get_occupation_part(input):
    return True if 'частичная занятость' in input else False
    
    
# Define get occupation feature function
def get_occupation_project(input):
    return True if 'проектная работа' in input else False


# Define get occupation feature function
def get_occupation_intern(input):
    return True if 'стажировка' in input else False 
    
     
# Define get occupation feature function
def get_occupation_volunteer(input):
    return True if 'волонтерство' in input else False  


# Define get schedule feature function
def get_schedule_full(input):
    return True if 'полный день' in input else False 


# Define get schedule feature function
def get_schedule_flex(input):
    return True if 'гибкий график' in input else False 
    
    
# Define get schedule feature function
def get_schedule_remote(input):
    return True if 'удалённая работа' in input else False 


# Define get schedule feature function
def get_schedule_watch(input):
    return True if 'вахтовый метод' in input else False 


# Define get schedule feature function
def get_schedule_shift(input):
    return True if 'сменный график' in input else False 

    
# Make new features        
hh_df['полная занятость'] = hh_df['Занятость'].apply(get_occupation_full)
hh_df['частичная занятость'] = hh_df['Занятость'].apply(get_occupation_part)
hh_df['проектная работа'] = hh_df['Занятость'].apply(get_occupation_project)
hh_df['стажировка'] = hh_df['Занятость'].apply(get_occupation_intern)
hh_df['волонтерство'] = hh_df['Занятость'].apply(get_occupation_volunteer)

hh_df['полный день'] = hh_df['График'].apply(get_schedule_full)
hh_df['гибкий график'] = hh_df['График'].apply(get_schedule_flex)
hh_df['удалённая работа'] = hh_df['График'].apply(get_schedule_remote)
hh_df['вахтовый метод'] = hh_df['График'].apply(get_schedule_watch)
hh_df['сменный график'] = hh_df['График'].apply(get_schedule_shift)

# Masks for data filtering
mask1 = (hh_df['проектная работа'] & hh_df['волонтерство'])
display(hh_df[mask1].shape[0])

mask2 = (hh_df['вахтовый метод'] & hh_df['гибкий график'])
display(hh_df[mask2].shape[0])

hh_df.drop(['Занятость', 'График'], axis= 1, inplace=True)
```

Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
ex_df = pd.read_csv('ExchangeRates.csv')
ex_df.drop(['per', 'time', 'vol'], axis= 1, inplace=True)

# Modify datetime formatting
ex_df['date'] = pd.to_datetime(ex_df['date']).dt.date
hh_df['date'] = pd.to_datetime(hh_df['Обновление резюме']).dt.date


# Define get salary function
def get_salary(input):
    return int(input.split()[0])

# Define get ISO currency code function
def get_currency_iso_code(salary):
    if 'бел' in salary:
        return 'BYN'
    if 'руб' in salary:
        return 'RUR'
    if 'грн' in salary:
        return 'UAH'
    if 'USD' in salary:
        return 'USD'
    if 'EUR' in salary:
        return 'EUR'
    if 'KGS' in salary:
        return 'KGS'
    if 'сум' in salary:
        return 'UZS'
    if 'AZN' in salary:
        return 'AZN'
    if 'KZT' in salary:
        return 'KZT'


# Make new features
hh_df['Зарплата(сумма)'] = hh_df['ЗП'].apply(get_salary)
hh_df['Зарплата(валюта)'] = hh_df['ЗП'].apply(get_currency_iso_code)

# Merge frames
hh_df = hh_df.merge(
    ex_df,
    left_on=['Зарплата(валюта)', 'date'],
    right_on=['currency', 'date'],
    how='left'
)

# Delete old features
hh_df.drop(['Обновление резюме', 'currency'], axis= 1, inplace=True)

# Make the dictionary for fill the empty cells
values = {
    'close': 1,
    'proportion': 1
}
# Fill the empty cells
hh_df = hh_df.fillna(values)

# Make new feature
hh_df['ЗП (руб)'] = (hh_df['Зарплата(сумма)']*hh_df['close'])/hh_df['proportion']

# Median salary in RUR
display(hh_df['ЗП (руб)'].median())

# Delete old features
hh_df.drop(['ЗП', 'Зарплата(сумма)', 'Зарплата(валюта)', 'close', 'proportion'], axis= 1, inplace=True)
```

Мы видим правостороннюю асимметрию распределения.

```python
# Define function for z-rejection algorithm
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    
    return outliers, cleaned


outliers, hh_df = outliers_z_score_mod(hh_df, 'Возраст', True, 3, 4)

display(outliers.shape[0])
display(outliers)
```
